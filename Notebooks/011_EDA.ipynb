````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Seleccionar las columnas de interés
columns_of_interest = [
    "start_time", "stop_time", "sold_quantity", "available_quantity", 
    "price", "base_price", "category_id"
]

filtered_df = df[columns_of_interest]

# Convertir 'start_time' y 'stop_time' a formato datetime
filtered_df['start_time'] = pd.to_datetime(filtered_df['start_time'], errors='coerce')
filtered_df['stop_time'] = pd.to_datetime(filtered_df['stop_time'], errors='coerce')

# Calcular la duración del listado en días
filtered_df['listing_duration'] = (filtered_df['stop_time'] - filtered_df['start_time']).dt.total_seconds() / (60 * 60 * 24)

# 1. Resumen estadístico de las variables
eda_summary = filtered_df.describe()

# 2. Verificar valores nulos
missing_values = filtered_df.isnull().sum()

# 3. Distribución de las variables numéricas
plt.figure(figsize=(14, 10))

# Sold Quantity
plt.subplot(2, 3, 1)
plt.hist(filtered_df['sold_quantity'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Distribución de Sold Quantity')

# Available Quantity
plt.subplot(2, 3, 2)
plt.hist(filtered_df['available_quantity'].dropna(), bins=30, color='green', edgecolor='black')
plt.title('Distribución de Available Quantity')

# Price
plt.subplot(2, 3, 3)
plt.hist(filtered_df['price'].dropna(), bins=30, color='red', edgecolor='black')
plt.title('Distribución de Price')

# Base Price
plt.subplot(2, 3, 4)
plt.hist(filtered_df['base_price'].dropna(), bins=30, color='purple', edgecolor='black')
plt.title('Distribución de Base Price')

# Listing Duration
plt.subplot(2, 3, 5)
plt.hist(filtered_df['listing_duration'].dropna(), bins=30, color='orange', edgecolor='black')
plt.title('Distribución de Listing Duration')

plt.tight_layout()
plt.show()

# Mostrar el resumen estadístico y valores nulos
print("Estadísticas descriptivas:")
print(eda_summary)

print("\nValores nulos por columna:")
print(missing_values)
````

### Conversión de la columna 'condition' en variable binaria

En este paso, se transforma la columna `condition` del DataFrame, que contiene valores categóricos ("new" y "used"), en una variable binaria. Esto facilita su uso en análisis numéricos y modelos de machine learning.

#### Pasos:

1. **Conversión a variable binaria**:
   - Se utiliza la función `apply()` junto con una expresión `lambda` para convertir los valores de la columna `condition` en una variable binaria. 
   - Si el valor es "new", la nueva columna `condition_binary` toma el valor **1**. Si el valor es "used", la columna toma el valor **0**.

2. **Verificación de los cambios**:
   - Se muestran las primeras filas de la nueva columna `condition_binary` junto con la columna original `condition` para verificar que la conversión se ha realizado correctamente.

#### Resultados:
- La columna `condition_binary` ahora representa la condición de los artículos de manera numérica:
  - **1** para artículos nuevos (`new`).
  - **0** para artículos usados (`used`).

````python
# Convertir la columna 'condition' en una variable binaria
df['condition_binary'] = df['condition'].apply(lambda x: 1 if x == 'new' else 0)

# Verificar los cambios
print(df[['condition', 'condition_binary']].head())
````

### Matriz de correlación con la variable binaria 'condition_binary'

En este análisis, se genera una matriz de correlación que incluye la variable `condition_binary` (que indica si un artículo es nuevo o usado) junto con otras variables numéricas del conjunto de datos. La correlación mide la relación lineal entre variables y puede tomar valores entre -1 y 1.

#### Pasos:

1. **Selección de columnas numéricas**:
   - Se seleccionan todas las columnas de tipo numérico (`float64` y `int64`) del DataFrame, incluidas aquellas relacionadas con los precios, cantidades y la variable binaria `condition_binary`.

2. **Cálculo de la matriz de correlación**:
   - Se calcula la matriz de correlación utilizando el método `corr()` de `pandas`, que devuelve los coeficientes de correlación entre todas las columnas numéricas. Un valor cercano a **1** indica una correlación positiva fuerte, mientras que un valor cercano a **-1** indica una correlación negativa fuerte.

3. **Visualización de la matriz de correlación**:
   - La matriz de correlación se visualiza utilizando un `heatmap` de `seaborn`, donde las correlaciones se representan en una escala de color que va de azul (correlación negativa) a rojo (correlación positiva). Los valores se anotan directamente sobre la gráfica para facilitar la interpretación.

#### Interpretación:
- **Correlación entre precios**: Se observa una fuerte correlación positiva entre `base_price`, `price`, y `original_price`, lo cual es esperable dado que estas variables están directamente relacionadas con los precios de los artículos.
- **Correlación de `condition_binary`**: La variable `condition_binary` muestra una baja correlación con otras variables numéricas, lo que sugiere que la condición de los artículos (nuevo o usado) no está fuertemente relacionada con otras variables como el precio o la cantidad disponible.
- **Duración del listado**: La variable `listing_duration` muestra una baja correlación con la mayoría de las otras variables, lo que indica que la duración de las publicaciones no está fuertemente influenciada por factores como el precio o la condición del artículo.

````python
# Incluir la columna binaria en la matriz de correlación
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Agregar la columna 'condition_binary' a la matriz de correlación
correlation_matrix = numerical_df.corr()

# Graficar la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de correlación con "condition_binary"')
plt.show()
````
